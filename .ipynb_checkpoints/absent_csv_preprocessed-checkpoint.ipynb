```python
import pandas as pd
import numpy as np
```

```python
absent_csv = pd.read_csv("Absenteeism_data.csv")
absent_csv
```

```python
absent_csv = absent_csv.drop(columns=["ID"])
absent_csv
```

```python
absent_csv["Reason for Absence"]
```

```python
# Create the categorical columns based on ranges
absent_csv['Reason_1'] = ((absent_csv['Reason for Absence'] >= 1) & (absent_csv['Reason for Absence'] <= 14)).astype(int)
absent_csv['Reason_2'] = ((absent_csv['Reason for Absence'] >= 15) & (absent_csv['Reason for Absence'] <= 17)).astype(int)
absent_csv['Reason_3'] = ((absent_csv['Reason for Absence'] >= 18) & (absent_csv['Reason for Absence'] <= 21)).astype(int)
absent_csv['Reason_4'] = ((absent_csv['Reason for Absence'] >= 22) & (absent_csv['Reason for Absence'] <= 28)).astype(int)

# Display the updated DataFrame
new_columns_order = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4'] + [col for col in absent_csv.columns if col not in ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']]

absent_csv = absent_csv[new_columns_order]

absent_csv = absent_csv.drop(columns=["Reason for Absence"])

print(absent_csv)
```
